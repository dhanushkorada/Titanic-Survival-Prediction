```python
import numpy as np
import pandas as pd
```

```python
train = pd.read_csv('train.csv')
```

```python
print(train)
```

```python
train = train.drop(['Name', 'Ticket','Cabin','PassengerId'], axis=1)
```

```python
y_train = train['Survived'].values
```

```python
print(y_train)
```

```python
x_train = train.drop(['Survived'], axis=1)
```

```python
nulls_per_column = x_train.isnull().sum()
print(nulls_per_column)
```

```python
mode_value = x_train['Embarked'].mode().iloc[0]
x_train['Embarked'].fillna(mode_value, inplace=True)
```

```python
mean_value = x_train['Age'].mean()
x_train['Age'].fillna(mean_value, inplace=True)
```

```python
nulls_per_column = x_train.isnull().sum()
print(nulls_per_column)
```

```python
print(x_train)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[1])],remainder='passthrough')
x_train = (ct.fit_transform(x_train))
```

```python
print(x_train)
```

```python
ct = ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[7])],remainder='passthrough')
x_train = (ct.fit_transform(x_train))
```

```python
print(x_train)
```

```python
print(x_train.shape)
```

```python
test = pd.read_csv('test.csv')
```

```python
print(test)
```

```python
test = test.drop(['Name', 'Ticket','Cabin','PassengerId'], axis=1)
```

```python
print(test)
```

```python
mean_value = test['Age'].mean()
test['Age'].fillna(mean_value, inplace=True)
```

```python
mean = test['Fare'].mean()
test['Fare'].fillna(mean, inplace=True)
```

```python
nulls_per_column = test.isnull().sum()
print(nulls_per_column)
```

```python
print(test)
```

```python
ct = ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[1])],remainder='passthrough')
test = (ct.fit_transform(test))
```

```python
print(test)
```

```python
ct = ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[7])],remainder='passthrough')
test = (ct.fit_transform(test))
```

```python
print(test)
```

```python
print(test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
x_train = sc_X.fit_transform(x_train)
test = sc_X.transform(test)
```

```python
from sklearn.svm import SVC
classifier = SVC(kernel = 'rbf', random_state = 0)
classifier.fit(x_train, y_train)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
y_pred = classifier.predict(test)
print(y_pred)
```

```python
passenger_id = np.arange(892, 892 + len(y_pred))

# Create the DataFrame
df = pd.DataFrame({
    'PassengerId': passenger_id,
    'Survived': y_pred
})
```

```python
print(df)
```

```python
print(df.shape)
```

```python
df.to_csv('prediction.csv',index=False)
```
